```python
import numpy as np
import matplotlib.pyplot as plt
from gudhi.representations import Landscape
import gudhi as gd
```

```python
#A single diagram with 2 points
D = np.array([[0, 10], [2, 8]])
```

```python
diags = [D]
l=Landscape(num_landscapes=2,resolution=20).fit_transform(diags)
print(l)
```

```python
X = np.linspace(np.min(D), np.max(D), 20)
plt.plot(X, l[0][0:20], color="blue")
plt.plot(X, l[0][20:], color="yellow")
plt.title("Persistence Landscape")
plt.xlabel("Time")
plt.ylabel("Landscape value")
plt.show()
```

```python
#calculate landscape value of a single protein
import os
import pandas as pd
```

```python
file = "corr/closed/1anf.corr_1.txt"
corr = pd.read_csv(file, header=None, delim_whitespace=True) 
dist = 1-np.abs(corr.to_numpy())
rips = gd.RipsComplex(distance_matrix=dist, 
                  max_edge_length=1.1)
simplexTree = rips.create_simplex_tree(max_dimension=2)
simplexTree.compute_persistence()
D = simplexTree.persistence_intervals_in_dimension(0)
D
```

```python
land=Landscape(num_landscapes=370,resolution=50)
```

```python
landD = land.fit_transform([D])
landD.shape
```
